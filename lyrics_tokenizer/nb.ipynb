```python
# tokenization

# we're gonna need to get rid of brackets indicating verses/choruses
# ^ parentheses too (indicating overlaid vocals)

# adapted from A2 honestly (credit: INFO 4300 staff)
def tokenize(text):
    txt = re.sub(r"(\[.*?\])|([,'()â€”?\-!\.\"])", "", text.lower())
    return [t for t in re.findall(r"[a-z]+", txt)]
```

```python
df['tknzd_lyrics'] = df['lyrics'].apply(tokenize)
```
