```python
import pandas as pd
import numpy as np
import re
```

```python
%ls
```

```python
df = pd.read_csv('lyrics_tokenizer/song_lyrics.csv')
```

```python
social_df = pd.read_csv('social_tags_df.csv')
```

```python
df.shape
```

```python
social_df = social_df.drop(columns='Unnamed: 0')
```

```python
big_df = df.merge(social_df, on=['title', 'artist'], how='inner')
```

```python
big_df = big_df.drop(columns = ['language_cld3','language_ft','language'])
```

```python
big_df
```

```python
def tokenize(text):
    txt = re.sub(r"(\[.*?\])|([,'()—?\-!\.\"])", "", text.lower())
    return [t for t in re.findall(r"[a-z]+", txt)]
```

```python
big_df['tknzd_lyrics'] = big_df['lyrics'].apply(tokenize)
```
